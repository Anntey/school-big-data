```python
from urllib import request
import json
import time

header = "date;time;vehicle;line;direction;latitude;longitude;speed"
url = "http://data.itsfactory.fi/journeys/api/1/vehicle-activity"
start_time = time.time()
rows_list = []

while(True):
    # -------------- Make request --------------
    res = urllib.request.urlopen(url).read()
    data = json.loads(res)
    # -------------- Append data to list (as one string per row) --------------
    num_buses = data["data"]["headers"]["paging"]["pageSize"]
    start_index = data["data"]["headers"]["paging"]["startIndex"]
    for i in range(start_index, num_buses):
        date = data["body"][i]["monitoredVehicleJourney"]["framedVehicleJourneyRef"]["dateFrameRef"]
        try:
            bus_time = data["body"][i]["recordedAtTime"][11:19]  
        except:
            bus_time = bus_time # annoying indexing/key/missing data bug, using approximation
            print("Something went wrong on")    
        line = data["body"][i]["monitoredVehicleJourney"]["lineRef"]
        vehicle = data["body"][i]["monitoredVehicleJourney"]["vehicleRef"]
        direction = data["body"][i]["monitoredVehicleJourney"]["directionRef"]
        latitude = data["body"][i]["monitoredVehicleJourney"]["vehicleLocation"]["latitude"]
        longitude = data["body"][i]["monitoredVehicleJourney"]["vehicleLocation"]["longitude"]
        speed = data["body"][i]["monitoredVehicleJourney"]["speed"]
        attributes = [date, bus_time, line, vehicle, direction, latitude, longitude, speed]
        row = ";".join(attributes)
        rows_list.append(row)
    # -------------- Check stopping criterion --------------#
    now_time = time.time()
    if now_time > (start_time + 60*60):
        break
    else:
        time.sleep(5)
```

```python
data = "\n".join(rows_list)
data = header + "\n" + data

with open("busdata.csv", "w") as outfile:
    for line in data:
        outfile.write(line)
```
