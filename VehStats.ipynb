```python
import warnings
import pandas as pd
warnings.filterwarnings("ignore")

header = "vehicleClass;registrationDate;vehicleGroup;usageType;variant;version;deploymentDate;colour;numDoors;chassisType;cabinType;numSeats;weight;vehMaxLoad;regMaxLoad;length;width;height;fuelType;engineCapacity;power(kW);numCylinders;turbo;hybrid;hybridClass;make;model;transmission;numGears;name;brakeType;modelRegNumber;euroFuelType;municipality;co2;mileage(km);serialNumber;lineNumber"
header = header.split(";")

car_df = pd.read_csv("Tieliikenne Avoin Data 5.6.csv", sep = ";", encoding = "latin")
car_df.columns = header
car_df.head()
```

```python
car_df = car_df[car_df["vehicleClass"].isin(["M1", "M1G"])]
car_df
```

```python
car_df = car_df.loc[:, ["make", "deploymentDate", "mileage(km)", "fuelType", "co2"]]
car_df
```

```python
car_df["deploymentDate"] = car_df["deploymentDate"].apply(lambda x: str(x)[0:4])
car_df
```

```python
car_df["deploymentDate"] = pd.to_datetime(car_df["deploymentDate"], format = "%Y", errors = "coerce") 
car_df
```

```python
car_df["mileage(km)"] = car_df["mileage(km)"].abs()
car_df
```

```python
car_df["fuelType"] = pd.to_numeric(car_df["fuelType"], errors = "coerce")
car_df
```

```python
car_df.isna().sum()
```

```python
car_df = car_df.dropna(axis = 0)
car_df
```

```python
makes = car_df["make"].value_counts()

top_makes = pd.DataFrame(makes[0:5])
top_makes.reset_index(level = 0, inplace = True)
top_makes.columns = ["make", "count"]

other_makes = pd.DataFrame({"make": "other", "count": [makes[5:-1].sum()]})

makes = pd.concat([top_makes, other_makes], axis = 0)
makes
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5, 5))
plt.pie(makes["count"], labels = makes["make"])
plt.show()
```

```python
miles = car_df[car_df["mileage(km)"] <= 1e6]
plt.hist(miles["mileage(km)"], bins = 1_000)
plt.show()
```

```python
def get_age(date):
    diff = 2019 - date.year
    if diff <= 5:
        res = "0-5"
    elif diff < 10:
        res = "5-10"
    elif diff < 15:
        res = "10-15"
    elif diff < 20:
        res = "15-20"
    else:
        res = "20-"
    return res

ages = car_df["deploymentDate"].apply(get_age)
ages = ages.value_counts()
ages
```

```python
plt.bar(ages.index[[3, 2, 0, 1, 4]], ages[[3, 2, 0, 1, 4]])
plt.show()
```

```python
emissions = car_df[["fuelType", "co2"]]
emissions_1 = emissions[emissions["fuelType"] == 1.0]["co2"]
emissions_2 = emissions[emissions["fuelType"] == 2.0]["co2"]

fig, ax1 = plt.subplots()
ax1.boxplot([emissions_1, emissions_2], showfliers = False, labels = ["gasoline", "diesel"])
plt.ylabel("CO2 emissions")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize = (9, 7))
# --------------------------------
axs[0, 0].pie(makes["count"], labels = makes["make"])
# --------------------------------
axs[0, 1].hist(miles["mileage(km)"], bins = 1_000)
axs[0, 1].set_ylabel("Count")
# --------------------------------
axs[1, 0].bar(ages.index[[3, 2, 0, 1, 4]], ages[[3, 2, 0, 1, 4]])
axs[1, 0].set_ylabel("Count")
# --------------------------------
axs[1, 1].boxplot([emissions_1, emissions_2], showfliers = False, labels = ["gasoline", "diesel"])
axs[1, 1].set_ylabel("CO2 emissions")
# --------------------------------
plt.tight_layout()
plt.savefig("vehstats.png")
plt.show()
```
